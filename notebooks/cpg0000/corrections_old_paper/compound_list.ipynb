```python
import pandas as pd
```

```python
data_index = pd.read_csv("/lustre/groups/ml01/workspace/alessandro.palma/imCPA_official/data/bbbc021_unannotated/processed/bbbc021_unannotated_large/metadata/bbbc021_unannotated_large.csv")
```

```python
to_save_csv = {}
```

```python
for cpd, dose in zip(data_index.CPD_NAME, data_index.DOSE):
    if cpd not in to_save_csv:
        to_save_csv[cpd] = []
        
    else:
        if dose not in to_save_csv[cpd]:
            to_save_csv[cpd].append(dose)
     
```

```python
for d in to_save_csv:
    to_save_csv[d] = [str(i) for i in to_save_csv[d]]
    print(d[0].upper()+d[1:], "&", ",".join(to_save_csv[d]), r"\\")
```
