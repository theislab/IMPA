```python
from IMPA.dataset.data_loader import CellDataLoader
from IMPA.solver import IMPAmodule
from omegaconf import OmegaConf
from tqdm import tqdm
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
import os
import torch
from tqdm import tqdm
import pandas as pd
import seaborn as sns
from skimage import io, color, filters, measure
import yaml
```

```python
path_to_configs = Path("/home/icb/alessandro.palma/environment/IMPA/IMPA/config_hydra/config")
```

**Read yaml files**

```python
def initialize_model_and_compute_memory(yaml_config):
    args_rdkit = OmegaConf.create(yaml_config)
    dataloader_rdkit = CellDataLoader(args_rdkit)
    solver_rdkit = IMPAmodule(args_rdkit, ".", dataloader_rdkit)
    return solver_rdkit

def count_params(model):
    print(sum(p.numel() for p in model.parameters()))
```

**BBBC021**

```python
with open(path_to_configs / 'bbbc021_all_retrain.yaml', 'r') as IMPA_bbbc021:
    # Load YAML data using safe_load() from the file
    yaml_IMPA_bbbc021 = yaml.safe_load(IMPA_bbbc021)

IMPA_bbbc021 = initialize_model_and_compute_memory(yaml_IMPA_bbbc021)
count_params(IMPA_bbbc021)
```

```python
with open(path_to_configs / 'bbbc021_starganv2_all_retrain.yaml', 'r') as starganv2_bbbc021:
    # Load YAML data using safe_load() from the file
    yaml_stargan_bbbc021 = yaml.safe_load(starganv2_bbbc021)

starganv2_bbbc021 = initialize_model_and_compute_memory(yaml_stargan_bbbc021)
count_params(starganv2_bbbc021)
```
